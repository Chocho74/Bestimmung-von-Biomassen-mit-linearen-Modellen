```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


data = pd.read_csv(r'G:\Meine Ablage\Studium\Synchronisation\10. Semester\Statistische_Verfahren\Projekt\buche.csv')
data=data.loc[data['author'] == 'Joosten']
data=np.log(data[['age','dbh','height','biom']])


breaks = np.log(np.array([0, 15, 35, 80 ]))

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(10, 6))

# Loop through subplots and create scatter plots with linear regression lines
for i, ax in enumerate(axs):
    variable_range = data[(data['dbh'] >= breaks[i]) & (data['dbh'] < breaks[i + 1])]
    x = variable_range['height']
    y = variable_range['biom']
    print(np.shape(variable_range))
    # Scatter plot
    ax.scatter(x, y, label='Data points')

    # Linear regression
    model = LinearRegression()
    model.fit(x.values.reshape(-1, 1), y)
    y_pred = model.predict(x.values.reshape(-1, 1))
    print(model.coef_)
    coef=np.round(model.coef_[0],2)
    ax.plot(x, y_pred, color='red', label='Linear regression')
    ax.text(0.5,8,'The slope is: ' + str(coef))

    ax.set_title(f'dbh range: {np.round(np.exp(breaks[i]),1)} - {np.round(np.exp(breaks[i + 1]),1)}')
    ax.set_xlabel('heihgt')
    ax.set_ylabel('biomass')
    ax.legend()
    ax.set_ylim((-1.5,10))
    ax.set_xlim((0.3,4.5))
    

plt.tight_layout()
plt.show()
```
